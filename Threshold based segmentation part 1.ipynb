```python
blurred_image = skimage.filters.gaussian(gray_image, sigma = 10)  #blurs the image to de-noise
print(blurred_image)
plt.imshow(blurred_image)
```

```python
fig, ax = plt.subplots()                               #returns a tuple containing a figure and axes object(s)
plt.imshow(blurred_image, cmap = 'gray')
```

```python
histogram, bin_edges = np.histogram(blurred_image, bins = 256, range = (0.0,1.0))
print(histogram)

fig, ax = plt.subplots()
plt.plot(bin_edges[0:-1],histogram)
plt.title("Grayscale Histogram")
plt.xlabel("Grayscale value")
plt.ylabel("Pixels")
plt.xlim(0,1.0)
```

```python
t = 0.9 #create a mask based on the threshold
binary_mask = blurred_image < t
fig, ax = plt.subplots()
plt.imshow(binary_mask, cmap = "gray")
```

```python
selection = image.copy()        #use the binary_mask to select the interesting part of the image
selection[~binary_mask] = 0

fig, ax = plt.subplots()
plt.imshow(selection)
plt.imshow(selection)
```
